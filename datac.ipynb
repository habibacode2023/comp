```python
import pandas as pd
inventory=pd.read_csv("inventory.csv")
new_column_names = ['Product_Name', 'Year_Month', 'Warehouse_Inventory','Inventory_Cost_Per_Unit']
inventory.columns = new_column_names
inventory.info()
```

```python
inventory.head()
```

```python
inventory['Year_Month'] = pd.to_datetime(inventory['Year_Month'], format='%Y%m').dt.strftime('%Y-%m')
inventory.head()
```

```python
df=inventory
def calculate_inventory_sum(df, years):
    # Convert 'Year_Month' column to string type
    df['Year_Month'] = df['Year_Month'].astype(str)
    
    inventory_sums = {}
    
    for year in years:
        # Filter the DataFrame based on the desired year
        filtered_df = df[df['Year_Month'].str.startswith(str(year))]
        
        # Calculate the sum of warehouse inventory
        inventory_sum = filtered_df['Warehouse_Inventory'].sum()
        
        # Store the inventory sum in the dictionary
        inventory_sums[year] = inventory_sum
    
    return inventory_sums


# Usage example
# Assuming the data is stored in a DataFrame named 'df'
# Please make sure you have pandas library installed

years_to_calculate = [2015, 2016, 2017]

inventory_sums = calculate_inventory_sum(df, years_to_calculate)

for year, inventory_sum in inventory_sums.items():
    print(f"Sum of warehouse inventory for {year}: {inventory_sum}")
```

```python
def calculate_inventory_cost_sum(df):
    inventory_cost_sums = {}

    # Extract the year from the 'Year_Month' column
    df['Year'] = df['Year_Month'].dt.year

    # Group the DataFrame by product name and year
    grouped_df = df.groupby(['Product_Name', 'Year'])

    # Iterate over each group
    for (product_name, year), group in grouped_df:
        # Calculate the sum of inventory cost per unit for the product and year
        inventory_cost_sum = group['Inventory_Cost_Per_Unit'].sum()

        # Store the inventory cost sum in the dictionary
        if product_name not in inventory_cost_sums:
            inventory_cost_sums[product_name] = {}
        inventory_cost_sums[product_name][year] = inventory_cost_sum
    
    return inventory_cost_sums


# Usage example
# Assuming the data is stored in a DataFrame named 'df'
# Please make sure you have pandas library installed

# Convert the 'Year_Month' column to datetime format
df['Year_Month'] = pd.to_datetime(df['Year_Month'])

inventory_cost_sums = calculate_inventory_cost_sum(df)

for product_name, year_data in inventory_cost_sums.items():
    print(f"Inventory cost per unit for {product_name}:")
    for year, inventory_cost_sum in year_data.items():
        print(f"- Year {year}: {inventory_cost_sum}")
```
